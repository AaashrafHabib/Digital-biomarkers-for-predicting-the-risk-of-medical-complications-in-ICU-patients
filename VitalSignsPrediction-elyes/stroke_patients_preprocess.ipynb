- dropping clinically invalid values

```python
# SOURCE https://github.com/HPI-CH/mimic-alarms/blob/main/create_clean_chartevents.ipynb
# Clinically valid value ranges
# Heart Rate (220045): 0-350
# Non Invasive Blood Pressure systolic (220179): 0-375
# Non Invasive Blood Pressure diastolic (220180): 0-375
# O2 saturation pulseoxymetry (220277): 0-100
# Source for Respiratory Rate https://journals.physiology.org/doi/abs/10.1152/jappl.1957.11.1.79#:~:text=Abstract,was%20in%20the%20expiratory%20duct.
# Respiratory rate 0-196
def drop_outliers(df):  
    df.loc[~(
         (df['Heart Rate'] < 0) | 
         (df['Non Invasive Blood Pressure systolic'] < 0)|
        (df['Non Invasive Blood Pressure diastolic']<0)|
               (df['Respiratory Rate']<0)),
        :] 
    df=df.loc[~(
        (df['Heart Rate'] > 350) | 
        (df['Non Invasive Blood Pressure systolic'] > 375)|
        (df['Non Invasive Blood Pressure diastolic'] > 375)|
        (df['Respiratory Rate']>196)),
        :] 
    return df
```

### Implementing different imputation techniques

**Imputation before or after splitting into train and test?**

You should split before pre-processing or imputing.

The division between training and test set is an attempt to replicate the situation where you have past information and are building a model which you will test on future as-yet unknown information: the training set takes the place of the past and the test set takes the place of the future, so you only get to test your trained model once.

Keeping the past/future analogy in mind, this means anything you do to pre-process or process your data, such as imputing missing values, you should do on the training set alone. You can then remember what you did to your training set if your test set also needs pre-processing or imputing, so that you do it the same way on both sets.

Added from comments: if you use the test data to affect the training data, then the test data is being used to build your model, so it ceases to be test data and will not provide a fair test of your model. You risk overfitting, and it was to discourage this that you separated out the test data in the first place

```python
################### STATIC FEATURES ######################
def static_impute(df):
    cols=["Height (cm)","Admission Weight (Kg)"]
    for column in cols:
        median = df[column].median()
        df[column]=df[column].fillna(median)
    return df
    
```

```python
                                ###### DYNAMIC FEATURES ################
#Simple Methods
VS_COLS= ["Heart Rate","Respiratory Rate","Non Invasive Blood Pressure diastolic","Non Invasive Blood Pressure systolic","Non Invasive Blood Pressure mean"]

def median_imputation(df):
    for column in VS_COLS:
        median = df[column].median()
        df[column]=df[column].fillna(median)

def mean_imputation(df):
    for column in VS_COLS:
        mean = df[column].mean()
        df[column]=df[column].fillna(mean)
    

def hot_deck_imputation(df):
    for column in VS_COLS:
        # Create a copy of the column without missing values
        non_missing_values = df[column].dropna().values
        
        # Count the number of missing values in the column
        missing_count = df[column].isnull().sum()
        
        # Generate random indices to sample from non-missing values
        indices = np.random.randint(low=0, high=len(non_missing_values), size=missing_count)
        
        # Replace missing values with randomly sampled non-missing values
        df.loc[df[column].isnull(), column] = non_missing_values[indices]

#Advanced
#KNN Imputation
from sklearn.impute import KNNImputer

def knn_imputation(df, n_neighbors=5):
    imputer = KNNImputer(n_neighbors=n_neighbors)
    df[VS_COLS] = imputer.fit_transform(df[VS_COLS])
#     df_filled = pd.DataFrame(imputed_data, columns=df.columns)
    return df

#Multivariate Imputation by Chained Equation â€” MICE
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

def mice_imputation(df, max_iter=10, random_state=0):
    imputer = IterativeImputer(max_iter=max_iter, random_state=random_state)
    df[VS_COLS] = imputer.fit_transform(df[VS_COLS])
#     df_filled = pd.DataFrame(imputed_data, columns=df.columns)
    return df




```

- encoding missingness by creating a missing indicator column for each vital sign

```python
def create_missing_indicator(df, columns):
    for col in columns:
        indicator_col = f"{col}_Missing"  
        df[indicator_col] = np.where(df[col].isnull(), 1, 0) 
        
    return df
```

```python
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
def dynamic_scale(df):
    final_df=pd.DataFrame()
    cols=['Heart Rate',
                  'Non Invasive Blood Pressure diastolic', 'Non Invasive Blood Pressure mean',
                 'Non Invasive Blood Pressure systolic', 'Respiratory Rate']
    for hadm in tqdm(df['HADM_ID'].unique()):
        scaler=StandardScaler()
        #extract the dataframe
        hadm_df=df[df['HADM_ID']==hadm]
        #scale the float cols
        hadm_df[cols]=scaler.fit_transform(hadm_df[cols])
#         print(hadm_df)
        final_df = pd.concat([final_df,hadm_df])
    return final_df
```

### Creating train/obs/pred window
- for each admission create train and target seq 
- then add them to final X and y tensor 

**apply the window function for each unique admission and then append the created sequences to X and y tensors the length of X and y tensors should be the length of unique patients**

```python
res_count=res.groupby('HADM_ID')['SUBJECT_ID'].count().reset_index()
res_count.rename(columns={"SUBJECT_ID":'count'},inplace=True)
res_count[res_count['count']<11]
```

```python
#apply this function for eacxh unique admission
#so we won't mix timestamps from different admission in the same training sequence
def create_window(df):
    n_observation=4
    n_forecast=2
    n_target=2
    X,y=[],[]
    for adm_id in tqdm(df['HADM_ID'].unique()):
        data=df[df['HADM_ID']==adm_id]
        target_cols=["stroke"]
        target_df=data[target_cols]
    #     data.drop(target_cols,axis=1,inplace=True)
        for i in range(len(data)-8):
            X.append(data.iloc[i:i+n_observation,2:])
            y.append(target_df[i+n_observation+n_forecast:i+n_observation+n_forecast+n_target])
    return np.array(X),np.array(y)
```
